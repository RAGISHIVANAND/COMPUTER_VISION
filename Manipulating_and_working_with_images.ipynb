# ***ROTATING THE IMAGE***

```python
import skimage.io as io
from skimage.transform import rotate

# Load the image
image = io.imread("/content/drive/MyDrive/Colab Notebooks/kajol.jpg")

# Rotate the image by 90 degrees clockwise
rotated_image = rotate(image, angle=-90)

# Display the rotated image
io.imshow(rotated_image)
io.show()

# Save the rotated image
io.imsave("rotated_image.jpg", rotated_image)


```

Zoom in /out

```python
from PIL import Image

# Load the image
image = Image.open("/content/drive/MyDrive/kajol.jpg")

# Define the zoom factor
zoom_factor = 2  # Zoom factor of 2 (zoom in by doubling the size)

# Get the width and height of the image
width, height = image.size

# Calculate the new width and height after zooming in
new_width = width * zoom_factor
new_height = height * zoom_factor

# Calculate the crop box coordinates
left = (width - new_width) // 10
top = (height - new_height) // 6
right = (width + new_width) // 4
bottom = (height + new_height) // 3

# Crop the zoomed-in portion from the image
zoomed_image = image.crop((left, top, right, bottom))

# Resize the cropped image to the new dimensions
zoomed_image = zoomed_image.resize((new_width, new_height))

# Display the zoomed-in image
zoomed_image.show()

# Save the zoomed-in image
zoomed_image.save("zoomed_image.jpg")


```
